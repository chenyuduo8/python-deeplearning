## 编写程序实现坐标点的象限分类
### 准备数据
#### 生成训练数据
<font size=4.5>1-4象限各100各数据

```python
z1=[]
z2=[]
z3=[]
z4=[]
for i in range(100):
    x=np.random.randint(0,100)
    y=np.random.randint(0,100)
    z1.append((x,y))
    z2.append((-1*x,y))
    z3.append((-1*x,-1*y))
    z4.append((x,-1*y))
z1=np.array(z1)
z2=np.array(z2)
z3=np.array(z3)
z4=np.array(z4)
train_data=np.concatenate([z1,z2,z3,z4])
print(train_data.shape)
# train_data.reshape(-1,2)
# print(train_data.shape)
print(train_data[100])
```

```python
dd=np.vstack((z1,z2,z3,z4))
print(dd[301])
dd.shape
```

#### 准备训练数据的标签
<font size=4.5>（1，0，0，0）代表第一象限<p>
（0，1，0，0）代表第二象限<p>
（0，0，1，0）代表第三象限<p>
（0，0，0，1）代表第四象限<p>

```python
train_label=np.zeros((400,4))
for i in range(100):
    train_label[i][0]=1
for i in range(100,200):
    train_label[i][1]=1
for i in range(200,300):
    train_label[i][2]=1
for i in range(300,400):
    train_label[i][3]=1
    
```

### 初始化一个二层网络对象

```python
net = TwoLayerNet(input_size=2, hidden_size=50, 
                  output_size=4)
x=np.array([2,1])
T=np.array([1,0,0,0])
y=net.predict(x)
print(f'预测结果：输入的点{x}在{y.argmax()+1}象限')
print(y)
net.loss(x,T)
```

```python
acc=net.accuracy(train_data,train_label)
print(f'训练前网络的预测精度为{acc*100}%')
```

### 训练网络

```python
np.random.shuffle(indices)

print(indices)
```

```python
data=np.zeros_like(train_data)
label=np.zeros_like(train_label)
j=0
for i in indices:    
    data[j] =train_data[i]
    label[j] = train_label[i]
    j+=1
```

```python
print(data[301],label[301])
```

### 网络训练（学习）

```python
train_loss_list = []
iters_num = 500  #学习次数
train_size = data.shape[0]
batch_size = 10
learning_rate = 0.01
```

```python
%%time

for i in range(iters_num):#执行学习次数
    grad = net.gradient(data,label) # 高速版!
    # 更新参数
    for key in ('W1', 'b1', 'W2', 'b2'):
        net.params[key] -= learning_rate * grad[key]  #梯度下降算法

    # 记录学习过程
    loss = net.loss(data, label)
    train_loss_list.append(loss)
    
```

#### 网络学习（训练）
机器学习的目标是找到使损失函数最小化的参数。在神经网络学习中，梯度法被用来沿梯度方向逐渐更新参数以最小化损失函数。数学上，参数更新可以表示为：
$x_0 = x_0 - \eta \frac{\partial f}{\partial x_0}\\
x_1 = x_1 - \eta \frac{\partial f}{\partial x_1}$

这里，$\eta$ 是学习率，它决定了参数更新的步长。

梯度下降有三种主要变种：
1. **随机梯度下降 (SGD)**: 每看一个数据就计算损失函数，然后求梯度更新参数。这种方法速度快，但可能在最优点附近震荡。
2. **批梯度下降**: 遍历整个数据集来计算损失函数和梯度，然后更新参数。这种方法计算量大，速度慢。
3. **小批量梯度下降 (Mini-batch)**: 这是一种折衷方法，它将数据分成若干批，按批更新参数。这减少了随机性，同时保持了计算效率。

在小批量梯度下降中，可以使用 `np.random.choice()` 从训练数据中随机选择一部分数据作为 mini-batch。

#### 网络训练


```python
train_loss_list = []

# 超参数
iters_num = 5000  #学习次数
train_size = d_train.shape[0]
batch_size = 100
learning_rate = 0.01
```

```python
%%time 

for i in range(iters_num):
    # 获取mini-batch
    batch_mask = np.random.choice(train_size, batch_size)
    d_batch = nd_train[batch_mask]
    l_batch = np.array(nl_train)[batch_mask]

    # 计算梯度
#     grad = network.numerical_gradient(x_batch, t_batch)
    grad = net.gradient(d_batch, l_batch) # 高速版!

    # 更新参数
    for key in ('W1', 'b1', 'W2', 'b2'):
        net.params[key] -= learning_rate * grad[key]

    # 记录学习过程
    loss = net.loss(d_batch, l_batch)
    train_loss_list.append(loss)
```

## 机器学习中的几个术语
<font size=4.5> &emsp;&emsp;**epoch**:一个epoch表示所有的数据送入网络中， 完成了一次前向计算 + 反向传播的过程。由于一个epoch 常常太大， 分成 几个小的 baches。将所有数据迭代训练一次是不够的， 需要反复多次才能拟合、收敛。<p>
&emsp;&emsp;在实际训练时、 将所有数据分成多个batch ， 每次送入一部分数据。使用单个epoch 更新权重不够。
随着epoch 数量的增加， 权重更新迭代的次数增多， 曲线从最开始的不拟合状态， 进入优化拟合状态， 最终进入过拟合。
epoch 如何设置： 大小与数据集的多样化程度有关， 多样化程度越强， epoch 越大。

&emsp;&emsp;**batchsize**
每个batch 中： 训练样本的数量。batch size 大小的选择也很重要， 最优化网络模型的性能+速度。当数据量较小， 计算机可以承载只有1个batch 的训练方式时， 收敛效果会好。

&emsp;&emsp;mini-batch : 将所有数据分为若干个batch , 每个batch 包含一部分训练样本。<p></p>
&emsp;&emsp;**iterations** 完成一次epoch 需要的batch 个数

&emsp;&emsp;batch numbers 就是 iterations .
</font>

```python
train_loss_list1 = []
train_acc_list = []
test_acc_list = []

# 超参数
iters_num = 10000
batch_size = 1000
learning_rate = 0.1
# 平均每个epoch的重复次数
iter_per_epoch = max(train_size / batch_size, 1)

# net1 = TwoLayerNet(input_size=784, hidden_size=50, output_size=10)

for i in range(iters_num):
    # 获取mini-batch
    batch_mask = np.random.choice(train_size, batch_size)
    x_batch = nd_train[batch_mask]
    t_batch = nl_train[batch_mask]

    # 计算梯度
#     grad = network.numerical_gradient(x_batch, t_batch)
    grad = net.gradient(x_batch, t_batch) # 高速版!

    # 更新参数
    for key in ('W1', 'b1', 'W2', 'b2'):
        net.params[key] -= learning_rate * grad[key]

    loss = net.loss(x_batch, t_batch)
    train_loss_list1.append(loss)
    # 计算每个epoch的识别精度
    if i % iter_per_epoch == 0:
        train_acc = net.accuracy(nd_train, nl_train)
        test_acc = net.accuracy(nd_test, nl_test)
        train_acc_list.append(train_acc)
        test_acc_list.append(test_acc)
        print("train acc, test acc | " + str(train_acc) + ", " + str(test_acc))
```

## 与学习相关的技巧--参数的更新
求 
$f(x,y)=\frac{1}{20}x^2+y^2$
的最小值为例，对比不同参数更新方法的结果,初始点(-7,2)

```python
def f(x,y):
    return x**2/20+y**2

def df(x,y):
    return x/10,2*y
x=np.arange(-10,10,0.25)
y=np.arange(-5,5,0.25)

X, Y = np.meshgrid(x, y)
z=f(X,Y)
plt.contour(X,Y,z)
plt.grid()
```

### SGD

$w=w-\eta * \frac{\partial l}{\partial w}$

```python
class SGD:

    """随机梯度下降法（Stochastic Gradient Descent）"""

    def __init__(self, lr=0.01):
        self.lr = lr
        
    def update(self, params, grads):
        for key in params.keys():
            params[key] -= self.lr * grads[key] 

```

```python
sgd=SGD(lr=0.95)
x_history=[]
y_history=[]
init_pos = (-7.0, 2.0)
params = {}
params['x'], params['y'] = init_pos[0], init_pos[1]
grads = {}
grads['x'], grads['y'] = 0, 0
for i in range(100):
    x_history.append(params['x'])
    y_history.append(params['y'])
    grads['x'],grads['y']=df(params['x'],params['y'])
    sgd.update(params,grads)
plt.plot(x_history,y_history,'o-')  ;  
plt.contour(X, Y, z)
plt.ylim(-5, 5)
plt.xlim(-10, 10)
plt.plot(0, 0, '+')
plt.title('Momentum')
plt.xlabel("x")
plt.ylabel("y")    
```

<font size=4.5>**基于 SGD 的最优化的更新路径：呈“之”字形朝最小值 (0, 0) 移动，效率低**。原因：就是 y 轴方向的坡度大，而 x 轴方向的坡度小。这里需要注意的是，虽然式 函数的最小值在 (x, y) = (0, 0) 处，但是梯度在很多地方并没有指向 (0, 0)。

### Momentum
<font size=4.5>Momentum 是“动量”的意思，和物理有关。用数学式表示 Momentum 方法，如下所示。
    $$
    v=av-\eta *\frac{\partial l}{\partial w}(1)\\
    w=w+v
    $$
    和前面的 SGD 一样，$w$ 表示要更新的权重参数，$\frac{\partial L}{\partial W} $表示损失函数关于$w$  的梯度，η 表示学习率。这里新出现了一个变量 v，对应物理上的速度。式（1）表示了物体在梯度方向上受力，在这个力的作用下，物体的速度增加这一物理法则。

```python
class Momentum:

    """Momentum SGD"""

    def __init__(self, lr=0.01, momentum=0.9):
        self.lr = lr
        self.momentum = momentum
        self.v = None
        
    def update(self, params, grads):
        if self.v is None:
            self.v = {}
            for key, val in params.items():                                
                self.v[key] = np.zeros_like(val)
                
        for key in params.keys():
            self.v[key] = self.momentum*self.v[key] - self.lr*grads[key] 
            params[key] += self.v[key]
```

```python
Mom=Momentum(lr=0.05)

x_history=[]
y_history=[]
init_pos = (-7.0, 2.0)
params = {}
params['x'], params['y'] = init_pos[0], init_pos[1]
grads = {}
grads['x'], grads['y'] = 0, 0
for i in range(100):
    x_history.append(params['x'])
    y_history.append(params['y'])
    grads['x'],grads['y']=df(params['x'],params['y'])
    Mom.update(params,grads)
plt.plot(x_history,y_history,'o-')  ;  
plt.contour(X, Y, z)
plt.ylim(-5, 5)
plt.xlim(-10, 10)
plt.plot(0, 0, '+')
plt.title('Momentum')
plt.xlabel("x")
plt.ylabel("y")    
```

### AdaGrad
<font size=4.5>在神经网络的学习中，学习率（数学式中记为 η）的值很重要。学习率过小，会导致学习花费过多时间；反过来，学习率过大，则会导致学习发散而不能正确进行。<p>
&emsp;&emsp;在关于学习率的有效技巧中，有一种被称为学习率衰减（learning rate decay）的方法，即随着学习的进行，使学习率逐渐减小。实际上，一开始“多”学，然后逐渐“少”学的方法，在神经网络的学习中经常被使用。<p>
&emsp;&emsp;逐渐减小学习率的想法，相当于将“全体”参数的学习率值一起降低。而 AdaGrad 进一步发展了这个想法，针对“一个一个”的参数，赋予其“定制”的值。<p>
&emsp;&emsp;AdaGrad 会为参数的每个元素适当地调整学习率，与此同时进行学习（AdaGrad 的 Ada 来自英文单词 Adaptive，即“适当的”的意思）。用数学式表示 AdaGrad 的更新方法:
   $$   
   h=h+\frac{\partial L}{\partial w}\odot \frac{\partial L}{\partial w}\quad (1)\\
 W= W-\eta\frac{1}{\sqrt h}\frac{\partial L}{\partial {W}}\quad(2)
    $$

```python
class AdaGrad:

    """AdaGrad"""

    def __init__(self, lr=0.01):
        self.lr = lr
        self.h = None
        
    def update(self, params, grads):
        if self.h is None:
            self.h = {}
            for key, val in params.items():
                self.h[key] = np.zeros_like(val)
            
        for key in params.keys():
            self.h[key] += grads[key] * grads[key]
            params[key] -= self.lr * grads[key] / (np.sqrt(self.h[key]) + 1e-7)
```

```python
ada=AdaGrad(lr=1.5)
x_history=[]
y_history=[]
init_pos = (-7.0, 2.0)
params = {}
params['x'], params['y'] = init_pos[0], init_pos[1]
grads = {}
grads['x'], grads['y'] = 0, 0
for i in range(100):
    x_history.append(params['x'])
    y_history.append(params['y'])
    grads['x'],grads['y']=df(params['x'],params['y'])
    ada.update(params,grads)
plt.plot(x_history,y_history,'o-')  ;  
plt.contour(X, Y, z)
plt.ylim(-5, 5)
plt.xlim(-10, 10)
plt.plot(0, 0, '+')
plt.title('AdaGrad')
plt.xlabel("x")
plt.ylabel("y")    
```

### Adam
<font size=4.5>
Adam 是 2015 年提出的新方法。它的理论有些复杂，直观地讲，就是融合了 Momentum 和 AdaGrad 的方法。通过组合前面两个方法的优点，有望实现参数空间的高效搜索。此外，进行超参数的“偏置校正”也是 Adam 的特征。

```python
class Adam:

    """Adam (http://arxiv.org/abs/1412.6980v8)"""

    def __init__(self, lr=0.001, beta1=0.9, beta2=0.999):
        self.lr = lr
        self.beta1 = beta1
        self.beta2 = beta2
        self.iter = 0
        self.m = None
        self.v = None
        
    def update(self, params, grads):
        if self.m is None:
            self.m, self.v = {}, {}
            for key, val in params.items():
                self.m[key] = np.zeros_like(val)
                self.v[key] = np.zeros_like(val)
        
        self.iter += 1
        lr_t  = self.lr * np.sqrt(1.0 - self.beta2**self.iter) / (1.0 - self.beta1**self.iter)         
        
        for key in params.keys():
            self.m[key] += (1 - self.beta1) * (grads[key] - self.m[key])
            self.v[key] += (1 - self.beta2) * (grads[key]**2 - self.v[key])
            
            params[key] -= lr_t * self.m[key] / (np.sqrt(self.v[key]) + 1e-7)
```

```python
ada=Adam(lr=0.3)
x_history=[]
y_history=[]
init_pos = (-7.0, 2.0)
params = {}
params['x'], params['y'] = init_pos[0], init_pos[1]
grads = {}
grads['x'], grads['y'] = 0, 0
for i in range(30):
    x_history.append(params['x'])
    y_history.append(params['y'])
    grads['x'],grads['y']=df(params['x'],params['y'])
    ada.update(params,grads)
plt.plot(x_history,y_history,'o-')  ;  
plt.contour(X, Y, z)
plt.ylim(-5, 5)
plt.xlim(-10, 10)
plt.plot(0, 0, '+')
plt.title('Adam')
plt.xlabel("x")
plt.ylabel("y")    
```
